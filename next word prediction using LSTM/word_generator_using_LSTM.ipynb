```python
import numpy as np
import pandas as pd
```

```python
import PyPDF2
```

```python
reader= PyPDF2.PdfReader('H:\\6th-sem\\Deep Learning models\\next word prediction using LSTM\\constitution of india questions.pdf')
```

```python
page_num = len(reader.pages)
page_num
```

```python
page = reader.pages
print(page[0].extract_text())
```

```python
page_num
```

```python
pdf_text = []
for i in range(page_num):
    pdf_text.append(page[i].extract_text())
```

```python
print(pdf_text)
```

```python
sentences = []

for paragraph in pdf_text:
    for sentence in paragraph.split('\n'):
        stripped_sentence = sentence.strip()
        if stripped_sentence:
            sentences.append(stripped_sentence)
```

```python
sentences[0:4]
```

```python
sentences = [sentence.replace('*', '').replace('-', '') for sentence in sentences]
```

```python
sentences[0:4]
```

```python
df = pd.read_csv('H:\\6th-sem\\Deep Learning models\\next word prediction using LSTM\\train.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
sentences = list(pd.array(df['Quotes']))
sentences[0:5]
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts(sentences)
```

```python
tokenizer.word_index
```

```python
len(tokenizer.word_index) #unique word
```

```python
tokenized_sentences = tokenizer.texts_to_sequences(sentences)
```

```python
tokenized_sentences[0:5]
```

```python
input_sequences =[]
for i in range(len(tokenized_sentences)):
  for j in range(len(tokenized_sentences[i])-1):
    n_gram = tokenized_sentences[i][:j+2]
    input_sequences.append(n_gram)
```

```python
input_sequences
```

```python
max_len_input= max([len(X) for X in input_sequences])
max_len_input
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences,maxlen = max_len_input, padding = 'pre')
```

```python
padded_input_sequences
```

```python
X_input = padded_input_sequences[:,:-1]
```

```python
Y_output = padded_input_sequences[:,-1]
```

```python
X_input[0:3]
```

```python
Y_output[0:3]
```

```python
X_input.shape
```

```python
Y_output.shape
```

```python
from tensorflow.keras.utils import to_categorical
Y_output = to_categorical(Y_output, num_classes = 600)
```

```python
Y_output
```

```python
Y_output.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
model_2 = Sequential()
model_2.add(Embedding(600, 100, input_length = 17))
model_2.add(LSTM(256))
model_2.add(Dense(600, activation = 'softmax'))
```

```python
model = Sequential()
model.add(Embedding(1199, 100, input_length = 34))
model.add(LSTM(518))
model.add(Dense(1199, activation = 'softmax'))
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer ='adam',
              metrics = ['accuracy'])
```

```python
model_2.compile(loss='categorical_crossentropy',
              optimizer ='adam',
              metrics = ['accuracy'])
```

```python
model_2.summary()
```

```python
model.summary()
```

```python
from keras.utils import plot_model
```

```python
plot_model(model ,show_shapes= True)
```

```python
model.fit(X_input, Y_output, epochs=100)
```

```python
model_2.fit(X_input, Y_output, epochs=100)
```
